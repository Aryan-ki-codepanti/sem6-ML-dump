```python
from sklearn.datasets import make_regression
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
```

```python
# here
from sklearn.linear_model import LinearRegression
```

```python

x = np.array([10,
15, 
20,     
25, 
30]).reshape(-1, 1)

y =  [1003,
1005,
1010,
1011,
1014]

reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_, reg.intercept_
```

```python
m = reg.coef_ # array
c = reg.intercept_
reg.predict([ [5]])
```

```python
# plotting
ax = plt.axes()
ax.scatter(x,y)
ax.set_xlabel("Temperature")
ax.set_ylabel("Pressure")
ax.axline((0, c) ,slope=m[0])
```
